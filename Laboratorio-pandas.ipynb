```python
import pandas as pd
```

# Ejercicio 1: Creación de DataFrame/Lista de Diccionarios
Crear una estructura de datos que contenga información sobre estudiantes (nombre, edad y calificación).

```python
# Python puro
students = [
{"name": "Alice", "age": 20, "grade": 85},
{"name": "Bob", "age": 21, "grade": 90},
{"name": "Charlie", "age": 19, "grade": 95}
]

for student in students:
  print(student)
```

```python
# Pandas
data = {
"name": ["Alice", "Bob", "Charlie"],
"age": [20, 21, 19],
"grade": [85, 90, 95]
}

df = pd.DataFrame(data)
print(df)
```

# Ejercicio 2: Selección de Columnas
Seleccionar la columna de calificaciones.

```python
# Python puro
grades = [student["grade"] for student in students]
print(grades)
```

```python
# Pandas
grades = df["grade"]
print(grades)
```

# Ejercicio 3: Filtrar Datos
Filtrar los estudiantes con una calificación mayor a 85.

```python
# Python puro
filtered_students = [student for student in
students if student["grade"] > 85]
print(filtered_students)
```

```python
# Pandas
filtered_df = df[df["grade"] > 85]
print(filtered_df)
```

# Ejercicio 4: Ordenar Datos
Ordenar los estudiantes por edad.

```python
# Python puro
sorted_students = sorted(students, key=lambda x: x["age"])
print(sorted_students)
```

```python
# Pandas
sorted_df = df.sort_values(by="age")
print(sorted_df)
```

# Ejercicio 5: Agregar Nueva Columna
Agregar una nueva columna que indique si el estudiante aprobó (calificación >= 60).

```python
# Python puro
for student in students:
  student["passed"] = student["grade"] >= 60
  print(student)
```

```python
# Pandas
df["passed"] = df["grade"] >= 60
print(df)
```
