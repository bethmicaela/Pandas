#### Si queremos hacer modificaciones a un df pero no queremos guardar los cambios directamente sobre el original, hacemos una copia con el comando .copy().
Defino un nuevo df que va a ser una copia del anterior y hay que especificar siempre que la copia es profunda, `porque normalmente cuando se copia un df en una variable, la variable no copia directamente la infomacion sino que copia la referencia a la memoria que se estan guardando los datos`

```
df = df_meteorites.copy(deep=True)
df.drop(['id'], axis=1, inplace=True)
```
