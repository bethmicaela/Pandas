```python
import pandas as pd
pd.__version__
```

```python
df_left = pd.DataFrame(
                      {'X': ['x0', 'x1', 'x2', 'x3'],
                       'W': ['w0', 'w1', 'w2', 'w3'], 
                       'Y': ['y0', 'y1', 'y2', 'y3']},
                       index=[0,1,2,3])
df_left
```

```python
df_right = pd.DataFrame(
                       {'Z': ['z2', 'z3', 'z4', 'z5'],
                         'A': ['a2', 'a3', 'a4', 'a5'], 
                         'Y': ['y2', 'y3', 'y4', 'y5']},
                        index=[2,3,4,5])
df_right
```

Hago un merge entre los dos df que tienen un parametro en comun (y2, y3):

```python
pd.merge(df_left,df_right)
```

Especificar el tipo de union que se esta presentando:

```python
pd.merge(df_left,df_right, how='inner', on='Y') # union(how) 'inner', columna(on) Y
```

Para unir dos columnas diferentes:

```python
pd.merge(df_left,df_right, how='inner', left_on='Mix', right_on='Y')
```

```python
pd.merge(df_left,df_right, how='inner', left_on='Mix', right_on='A')
```

### ***Cuantos tipos de merge podemos llegar a tener:***

![](http://www.datasciencemadesimple.com/wp-content/uploads/2017/09/join-or-merge-in-python-pandas-1.png)

`how='inner'`  Funciona trayendo los parametros que tienen en comun los df

`how='outer'`  Se juntan todos los parametros

`how='left'`  Se respeta el df de izquierda y se le añade los parametros del de la derecha

`how='right'`  Se respeta de df de la derecha y se le añade los parametros del de la izquierda





Como funcionan en detalle:

Con inner se van a analizar solo en los variables que tienen en comun

```python
pd.merge(df_left, df_right, how='inner', on='Y')
```

Con left el df de izquirda se mantiene intacto y se le añade los datos del df derecho

```python
pd.merge(df_left, df_right, how='left', on='Y')
```

Con right se respeta el tamaño y los numeros del df de la derecha y se le añaden los datos de la iquierda

```python
pd.merge(df_left, df_right, how='right', on='Y')
```

Para mantener todos los datos, tanto izquierdos como derechos, lo hago a travez de outer:

```python
pd.merge(df_left, df_right, how='outer', on='Y')
```
