```python
dict_data = {
    'edad' :     [ 10, 9, 13, 14, 12, 11, 12],
    'cm' : [ 115, 110, 130, 155, 125, 120, 125],
    'pais' :    [ 'co', 'mx', 'co', 'mx', 'mx', 'ch', 'ch'],
    'genero' :  [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'Q1' : [ 5, 10, 8, np.nan, 7, 8, 3],
    'Q2' : [ 7, 9, 9, 8, 8, 8, 9.]
}
dict_data
```

```python
df = pd.DataFrame(dict_data)
df
```

Para hacer enfasis en solo algunos estudiantes, uso la funcion loc(de localizar):

loc me permite seleccionar datos a travez de sus indices y columnas [indice/s, columna/s].

iloc me lo permite a travez de sus posicion dentro del dataframe empenzando de cero, [indice 0 1 2 3 etc(filas), columna 0 1 2 3 etc].

```python
df.iloc[2,1] 
```

Con iloc tambien se puede seleccionar varios conjuntos de elementos, ejemplo me traigo dos columnas:

```python
df.iloc[2,[1,3]] 
```

```python
df.iloc[[2,4,5],[1,3]] # [indices, columnas]
```

```python
df.iloc[:,[1,3]] # : (2 puntos) significa que me devuelva todo el indice completo o sea, toda la fila
```

Evaluo a las personas mayores o iguales a 12 anios, las filtro:

```python
df['edad']>=12 
```

Traigo subconjuntos de datos filtrados por la edad en donde las personas son mayores o iguales a 12 anios:

```python
df[df['edad']>=12] 
```

Combino varias condiciones separadas por parentesis y &, que sean mayores o iguales a 12 y que sean del pais mexico:

```python
df[(df['edad'] >= 12) & (df['pais'] == 'mx')] 
```

```python
df.query('edad>=12') # obtengo el mismo resultado que en la linea 39
```

```python
df['Q2'] >= df['Q1'] # obtengo valores booleanos
```
