### ***FUNCIONES***

Si quisieramos aplicar una funcion directamente, euros a dolares:

```python
def mean_eur2usd(x):
  return np.mean(x) * 1.12 # Que me retorne el promedio de mi variable y lo multiplique por 1.12
```

```python
mean_eur2usd(100) # A 100 se le calcula el promedio y se lo multiplica por 1.12
```

Si coloco una LISTA de elementos, sobre esos elementos se va a calcular un promedio de ellos y multiplicarse por 1.12

```python
mean_eur2usd([100, 102])
```

Volviendo a la funcion de groupby, lo que hago es aplicar la funcion mean_eur2usd

```python
df.groupby('sex')[['total_bill', 'prct_tip']].apply(mean_eur2usd)
```

Puedo agrupar mas variables, agrupo no solo por el genero, sino tambien por el tiempo

`Lunch -> almuerzo`

`Dinner -> cena`

```python
df.groupby(['sex','time'])[['total_bill', 'prct_tip']].apply(mean_eur2usd)
```

En general cuando usamos apply podemos colocar funciones definidas por nosotros o las mismas funciones definidas por numpy(por ej)

```python
df.groupby(['sex','time'])[['total_bill', 'prct_tip']].apply(np.std) # Si quiero calcular la desviacion estandar de los valores de esos platos, coloco np.std
```

Podemos definir nuestras propias funciones dentro de apply, usando el operador lambda y compiando la linea de mi anterior funcion donde retornaba una ecuacion

```python
df.groupby(['sex','time'])[['total_bill', 'prct_tip']].apply(lambda x: np.mean(x) * 1.12) # Obtengo el mismo resultado que la linea 19
```

En general puedo declarar una variable ***`y`*** para colocar dentro mi función y retornar el valor de esa variable

```python
def mean_eur2usd(x):
  y = np.mean(x) * 1.12
  return y
```

Si quisiese aplicar mas de una función, podemos utilizar la función aggregate

```python
df.groupby(['sex','time'])[['total_bill', 'prct_tip']].aggregate([np.mean, np.max]) # Dentro de una lista para aggregate
```

Podemos definir a través de aggregate funciones enfocadas en una sola categoría 

```python
df.groupby(['sex','time'])[['total_bill', 'prct_tip']].agg([np.mean, np.max])
```

### ***DICCIONARIOS***

Definimos un diccionarios de operaciones, contiene dos llaves que, tip calcula la propina minima y maxima y total bill calcula el promedio y el promedio en dolares

```python
dict_agg = {'tip':[min, max], 'total_bill':[np.mean, mean_eur2usd]}
dict_agg
```

```python
df.groupby(['sex','time'])[['total_bill', 'tip']].agg(dict_agg)
```

Puedo definir filtros: que me regrese el promedio de dolares y de esto escogo la columna total_bill y le digo que me calcule el promedio. Ademas quiero todas las facturas cuyo valor sean mayor a 20

```python
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
```

Creo un groupby

```python
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
# Obtengo un df parecido al original, pero este esta filtrado con los valores mayores a 20
# Entonces en vez de tener 244 elementos, tenemos 209
```

Para validar que nuestro filtro funcionó correctamente, mostrando en la fila de las mujeres 0(nan)

```python
df_filtered.groupby(['sex','time']).count()
```

```python
df
```

Creo una columna nueva, que contiene numeros 1

```python
df['ones'] = 1
df
```

Agrupo por genero y fumadores

```python
df_g = df.groupby(['sex', 'smoker']) [['ones']].sum()
df_g
```

Para convertir lo anterior a porcentaje
level=0, significa que me voy a posicionar en el nivel cero, o sea en el de genero  





```python
df_g.groupby(level=0).apply( # Aplico una función 
    
lambda x:
x / x.sum() * 100  # A cada uno de estos valores(x) que estan en el df anterior, lo divida en el valor promedio de su grupo y a eso lo multiplique por 100, para obtener el porcentaje

)


```

Otra forma de traer valores de nuestros datos es también tranformar nustras variables numéricas(continuas), en una variable categórica:

Esto lo hacemos con cut, haciendo enfasis en el total de las facturas y creando tres categorías con bins = 3

Para ver los valores de las categorias con value_counts():

vemos que, en la primer categoría de 3 a 18 dolares tenemos 140 valores, en la segunda de 18 a 34 dolares tenemos 88 valores y en la tercera de 34 a 50 dolares tenemos 16 valores

```python
df['bin_total'] = pd.cut(df['total_bill'], bins = [3,18,35,60])
df
```

También podemos definir el ancho de nuestras categorías colocando una lista:

```python
pd.cut(df['total_bill'], bins = [3,18,35,60]).value_counts()  # Me quedan dos categorías
# Son 125 valores entre 3 y 18 dolares, 101 valores entre 18 y 35, y 16 valores entre 35 y 60 dólares 🤑 
```

Sobre esta categoria tambien podemos hacer groupby

```python
df.groupby(['time', 'bin_total'])[['ones']].count()
```

L o convertimos a porcentaje con la formula de lambda

```python
df.groupby(['time', 'bin_total'])[['ones']].count().groupby(level=0).apply(
    
lambda x:
x / x.sum() * 100  

)
```

### ***PIVOT***

```python
df
```

Creamos un grupo de dos categorias, genero y hora de comida, y voy a calcular el promedio de las facturas

```python
df.groupby(['sex', 'time'])[['total_bill']].mean()
```

Borro los indices y los vuelvo columnas, y los guardo en una nueva variable(df_gp)

```python
df_gp = df.groupby(['sex', 'time'])[['total_bill']].mean().reset_index()
df_gp
```

Sobre mi nueva variable aplico una tabla dinamica(pivot_table)

Reestructuro mi tabla 

```python
df_gp.pivot_table(values='total_bill', index='sex', columns='time')
```

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

Si lo hago con df, obtengo el mismo resultado:

Pivot table internamente esta realizando el promedio de cada uno de los valores

```python
df.pivot_table(values='total_bill', index='sex', columns='time')
```

Hay una funcion que yo puedo manipular y cambiar, segun la necesidad de mi problema (agregate function), entonces en vez del promedio, me calcula la media(aggfunc=np.median)

```python
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=np.median)
```

Asi como puedo colocar una funcion, tambien puedo colocar un diccionario o una lista de funciones, como funcionan agregate y groupby.

Por ejemplo ahora quiero la desviacion estandar (std)

lo guado en una nueva variable como df_pivot

```python
df_pivot = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median, np.std])
df_pivot
```

Deshago mis categorias con unstack, las categorias anteriores que estaban totalmente agrupadas, pasan totalmente a ser filas

```python
df_pivot.unstack()
```

Si lo quiero volver a un df con reset_index()

Obtengo un df donde mis columnas previamente, ahora hacen parte de las filas

Tambien mis columnas median y std ahora aparecen en la columna level_o

```python
df_pivot.unstack(). reset_index()
```
