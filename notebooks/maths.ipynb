Variable tipo tiempo:

Cambio la variable texto/objeto de timestamp a variable tipo tiempo/fecha con pd.to_datetime

```python
df_lmerged['timestamp'] = pd.to_datetime(df_lmerged['timestamp'])
df_lmerged['timestamp']
```

Creo una nueva columna llamada 'hora':

```python
df_lmerged['hour'] = df_lmerged['timestamp'].dt.hour
df_lmerged['hour']
```

Quitar el tiempo del df:

`Me quedan solo variables de tipo numericas`

```python
df = df_lmerged.iloc[:, 1:] # Para que es lmerged?
df
```

## Funciones más complejas y lambdas:

x -> variable de entrada

```python
def fun_1(x):
  y = x**2 + 1
  return y
```

Dentro de la funcion(x) puedo colocar un numero y este hará la ecuacion que hay dentro y retornara el resultado

```python
fun_1(10)
```

```python
import numpy as np
```

Armo un rango de datos:

```python
np.arange(-5, 6)
```

Cuento la cantidad de datos que tiene:

```python
np.arange(-5, 6).shape
```

Cada uno de esos datos pasa por la funcion, elevandose al cuadrado y sumando 1

```python
fun_1(np.arange(-5, 6))
```

Se lo aplico al dataframe:

```python
df['hour'].apply(fun_1)
```

Puedo hacer funciones mas complejas agregando argumentos:

```python
def fun_2(x, a=1, b=0):
  y = x**2 + a*x + b
  return y
```

Le asigno nuevos argumentos:

```python
fun_2(10, a=20, b=-100)
```

Agrego esta nuevo funcion a mi df:

```python
df['hour'].apply(fun_2, args=(20, -100))
```

A args tambien los puedo llamar como a y b:

```python
df['hour'].apply(fun_2, a=20, b=-100)
```

Uso la funcion lambda:

Cambio la temperatura(t1) a grados kelvin

```python
df['t1'].apply(lambda x: x+273)
```

El promedio(mean) de cada una de las columnas, aplicando una funcion a todo el df:

```python
df.apply(lambda x: x.mean())
```

Ahora de las filas:

```python
df.apply(lambda x: x.mean(), axis=1)
```

La desviacion estandar:

```python
df.apply(lambda x: x.std(), axis=1)
```

Operaciones entre las columnas t1 y t2, y se hace a lo largo de las filas:

```python
df.apply(lambda x: x['t1'] - x['t2'], axis=1)
```

Opero sobre todos los valores de forma independiente:

En este caso divido a todos los valores de mi df por 1000

```python
df.applymap(lambda x: x/1000)
```
