### ***Concatenar matrices con numpy***

```python
np.concatenate([x1, x2])
```

```python
np.concatenate([x1, x2]).shape
```

Concatenar en el eje de las columnas especificandolo:

```python
np.concatenate([x1, x2], axis = 1)
```

Dimensiones del anterior:

```python
np.concatenate([x1, x2], axis = 1).shape
```

### ***Concatenar con Pandas***

Creo dos series:

```python
s1 = pd.Series(x1[0], index = ['a','b','c','d','e'])
s1
```

```python
s2 = pd.Series(x2[0], index = ['c','d','e','f','g'])
s2
```

Concateno las dos series a lo largo de las filas:


```python
pd.concat([s1, s2])
```

Concateno las dos series a lo largo de las columnas:

```python
pd.concat([s1, s2], axis=1) # Aparece NaN porque esas letras no tienen valor  y respetan los indices
```

Si quiero concatenar las series a los largo de las columnas pero sin respetar el indice:

```python
s1.reset_index(drop=True) # reseteo su indice
```

Obtengo ambas series concatenadas sin respetar el indice:

```python
pd.concat([s1.reset_index(drop=True), s2.reset_index(drop=True)], axis=1)
```

### ***Concatenar DataFrame***

```python
df1 = pd.DataFrame(np.random.rand(3,2)*10,columns=['a','b'])
df1
```

```python
df2 = pd.DataFrame(np.random.rand(3,2)*-1,columns=['a','b'], index = [2, 3, 4])
df2
```

Concateno los dos df:

```python
pd.concat([df1, df2])
```

Los concateno a lo largo de columnas:

```python
pd.concat([df1, df2], axis=1)
```

Si queremos hacer enfasis solo en el indice que se comparte(2):

```python
pd.concat([df1, df2], axis=1, join='inner')
```

Concatenar dos df sin respetar sus indices:

```python
pd.concat([df1.reset_index(drop=True),
           df2.reset_index(drop=True)
           ], axis=1)
```

Funcion append para unir dos df

```python
df1.append(df2)
```

Puedo hacer multiples append y se van a ir pegando uno abajo del otro:

```python
df1.append(df2).append(df2)
```

Para pegar los append a lo largo de las columnas con la traspuesta, lo que hace es invertir las filas por las columnas

```python
df1.T.append(df2.T).T
```
