### ***Ciclo for***

El for es una estructura de python que nos permite hacer bucles

```python
for i in [0, 1, 2, 3, 'hola']:  # Lista de elementos
  print(i)
```

Definimos la llave del groupby, el grupo de informacion que contiene `in` dentro del df:

-- Seguimos trabajando con el precio `grouped_price` y a cada uno de estos grupos voy a seleccionar de ellos la columna price y sobre ellos aplico el promedio

-- Imprimo que esta pasando en cada una de esas iteraciones `for` dado el tipo corte `cut` voy a pedirle cual es el precio promedio `price`

```python
for key_group, group in df.groupby('cut'):
  grouped_price = group['price'].mean()

  print('Cut: {}, Price: {} \n'.format(key_group, grouped_price))
```

Podemos agrupar usando varios parámetros, ésta vez usando el tipo de corte y color.  Sobre este podemos ver la influencia en el precio

Convertir a df con to_frame()


```python
df.groupby(['cut', 'color'])['price'].mean().to_frame()
```

### ***Aggregate***

Con aggregate defino una lista de las funciones que le quiero aplicar, por ejemplo si quiero que me calcule el mínimo promedio y maximo de np

```python
df.groupby(['cut', 'color'])['price'].aggregate(('min', np.mean,max))
```

### ***Con funciones***

Podemos definir funciones para hacer el mismo proceso

```python
def mean_kilo(x):
  return np.mean(x)/1000 # Retorno el promedio de una variable x y los divido en mil
```

Si a mi elemento anterior en aggregate, aplico mi funcion(lo agrego dentro de aggregate)

```python
df.groupby(['cut', 'color'])['price'].aggregate(('min', np.mean,max, mean_kilo)).head(10)
```

### ***Con diccionario***

Podemos hacer un diccionario de instrucciones que podemos colocar dentro de aggregate

para los quilates `carat` me calcule los valores minimos y maximos de un grupo

y para el precio `price` me calcule los valores promedio `mean` y tambien los  kilos promedio `mean_kilo`

Este diccionario de agregaciones tiene como llaves las columnas donde vamos a aplicar nuestras funciones y la lista de funciones que vamos a aplicar a cada una.

```python
dict_agg = {'carat':[min, max], 'price':[np.mean, mean_kilo]}
dict_agg
```

Traemos el group by junto con el aggregate y aplicamos el diccionario que creamos de funciones

Lo que obtenemos es que se agrupo bajo dos categorias, el corte y el color, lo cual me definen dos multiples indices y asi mismo le pedi que para las columnas de quilates, calcule el minimo y el maximo, y para el precio que calcule el promedio y el promedio dividido en mil

```python
df.groupby(['cut', 'color']).aggregate(dict_agg)
```

### ***Crear filtros***

```python
def f_filter(x):
  return mean_kilo(x['price']) > 4
```

```python
df.groupby('cut').filter(f_filter)
```

Para entender que esta pasando en la linea anterior, selecciono mi linea de corte y pongo `unique` y lo que sucede es que se filtraron dos categorias, premium & fair que son los tipos de corte que mas costo elevado tienen para los diamandtes

```python
df.groupby('cut').filter(f_filter)['cut'].unique()
```
